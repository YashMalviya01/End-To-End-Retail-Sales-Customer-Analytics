```python
# Importing all libraries

import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# Importing os and dataset into notebook

import os

BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), ".."))

sales = pd.read_csv(os.path.join(BASE_DIR, "Sales.csv"))
customer = pd.read_csv(os.path.join(BASE_DIR, "Customer.csv"))
product = pd.read_csv(os.path.join(BASE_DIR, "Product.csv"))
store = pd.read_csv(os.path.join(BASE_DIR, "Store.csv"))
date = pd.read_csv(os.path.join(BASE_DIR, "Date.csv"))
currency = pd.read_csv(os.path.join(BASE_DIR, "Currency.csv"))
```

Step 2 - 

Standardize Data 

Standardize the values so that it does not give any error in SQL.

```python
def clean_columns(df):
    df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
    return df

sales = clean_columns(sales)
customer = clean_columns(customer)
product = clean_columns(product)
store = clean_columns(store)
date = clean_columns(date)
currency = clean_columns(currency)
```

```python
# Fixing the Data Type for Sales

sales['orderdate'] = pd.to_datetime(sales['orderdate'])
sales['deliverydate'] = pd.to_datetime(sales['deliverydate'])
sales['quantity'] = sales['quantity'].astype(int)
sales['unitprice'] = sales['unitprice'].astype(float)
```

```python
# Fixing the date data type

date['date'] = pd.to_datetime(date['date'])
```

```python
# Dropping all duplicate values from all datasets

sales = sales.drop_duplicates()
customer = customer.drop_duplicates()
date = date.drop_duplicates()
currency = currency.drop_duplicates()
product = product.drop_duplicates()
store = store.drop_duplicates()
```

Step 3 - 

Calculating and Removing all Null Values from our Datasets. 

```python
# 1 For Sales
sales.isnull().sum()
```

```python
# 2 For Customers

customer.isnull().sum()
```

As we can see there are null values in company and state so we are filling them with (Unknown).

```python
customer['company'] = customer['company'].fillna('unknown')
customer['state'] = customer['state'].fillna('unknown')
customer[['company', 'state']].head(20)
```

```python
# 3 For Currency table

currency.isnull().sum()
```

```python
# 4 For date table

date.isnull().sum()
```

```python
# 5 Forc Product table

product.isnull().sum()
```

In this table there were null values in weight and weight unit columns. so we have filled weight with 0 and weightunit with unknown

```python
product[product['weight'].isnull()].head()
product[product['weightunit'].isnull()].head()
```

```python
product['weight'] = product['weight'].fillna(0)
product['weightunit'] = product['weightunit'].fillna('Unknown')
```

```python
store.isnull().sum()
```

```python
# 6 For stores table

store['status'].value_counts(dropna=False)
```

```python
# Therre were nan values in status section so we have filled it with open.

store['status'] = store['status'].fillna('Open')
```

```python
# It converts the `closedate` column into a proper datetime format.
# Invalid or unparseable values are automatically turned into `NaT` (missing date)


store['closedate'] = pd.to_datetime(store['closedate'], errors='coerce')
```

```python
# if status is Open → closedate should be NaT
store.loc[store['status'] == 'Open', 'closedate'] = pd.NaT
```

```python
store['squaremeters'] = store['squaremeters'].fillna(store['squaremeters'].median())
```

We will calculate the total revenue in sales table.

Step 5 - 

Saving the clean dataset so that we can perform further analysis


```python
from sqlalchemy import create_engine

engine = create_engine("postgresql+psycopg2://postgres:postgres123@localhost:5432/contoso")
```

```python
sales.to_sql("clean_sales", engine, if_exists="replace", index=False)
customer.to_sql("clean_customer", engine, if_exists="replace", index=False)
product.to_sql("clean_product", engine, if_exists="replace", index=False)
store.to_sql("clean_store", engine, if_exists="replace", index=False)
date.to_sql("clean_date", engine, if_exists="replace", index=False)
currency.to_sql("clean_currency", engine, if_exists="replace", index=False)
```
