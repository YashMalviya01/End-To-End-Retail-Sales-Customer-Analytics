```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("postgresql+psycopg2://postgres:postgres123@localhost:5432/contoso")

sales = pd.read_sql("SELECT * FROM feat_sales", engine)
customer = pd.read_sql("SELECT * FROM feat_customer", engine)
product = pd.read_sql("SELECT * FROM feat_product", engine)
store = pd.read_sql("SELECT * FROM feat_store", engine)
date = pd.read_sql("SELECT * FROM feat_date", engine)
currency = pd.read_sql("SELECT * FROM currency", engine)
```

```python
fact_sales = sales[[
    'orderkey',
    'linenumber',
    'orderdate',
    'deliverydate',
    'customerkey',
    'storekey',
    'productkey',
    'quantity',
    'unitprice',
    'netprice',
    'unitcost',
    'currencycode',
    'exchangerate',
    'salesamount',
    'delivery_days'
]]
```

```python
dim_customer = customer[[
    'customerkey',
    'full_name',
    'gender',
    'city',
    'state',
    'country',
    'age_group'
]]
```

```python
dim_product = product[[
    'productkey',
    'productname',
    'brand',
    'manufacturer',
    'categoryname',
    'price_category'
]]
```

```python
dim_store = store[[
    'storekey',
    'countryname',
    'state',
    'status',
    'store_age',
    'store_size',
    'squaremeters'
]]
```

```python
dim_date = sales[[
    'orderdate',
    'year',
    'month',
    'quarter',
    'day_name'
]].drop_duplicates()
```

```python
fact_sales.to_sql(
    "fact_sales",
    engine,
    if_exists="replace",
    index=False
)

dim_customer.to_sql(
    "dim_customer",
    engine,
    if_exists="replace",
    index=False
)

dim_product.to_sql(
    "dim_product",
    engine,
    if_exists="replace",
    index=False
)

dim_store.to_sql(
    "dim_store",
    engine,
    if_exists="replace",
    index=False
)

dim_date.to_sql(
    "dim_date",
    engine,
    if_exists="replace",
    index=False
)
```
