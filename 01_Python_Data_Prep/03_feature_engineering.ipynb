Feature Engineering 

Objective - 
Create business-focused analytical features from cleaned retail datasets.

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine(
    "postgresql+psycopg2://postgres:postgres123@localhost:5432/contoso"
)
```

Step 2 - 

Load Clean Datasets

```python
sales = pd.read_sql("SELECT * FROM clean_sales", engine)
customer = pd.read_sql("SELECT * FROM clean_customer", engine)
product = pd.read_sql("SELECT * FROM clean_product", engine)
store = pd.read_sql("SELECT * FROM clean_store", engine)
date = pd.read_sql("SELECT * FROM clean_date", engine)
currency = pd.read_sql("SELECT * FROM clean_currency", engine)
```

Step 3 - 

Feature Engineering for Sales.


Sales Feature
Create revenue and time-based analytical features.

```python
sales['orderdate'] = pd.to_datetime(sales['orderdate'])
sales['deliverydate'] = pd.to_datetime(sales['orderdate'])
```

```python
# revenue feature

sales['salesamount'] = sales['quantity'] * sales['unitprice']
sales['salesamount']
```

```python
#Time Features

# Time Features
sales['year'] = sales['orderdate'].dt.year
sales['month'] = sales['orderdate'].dt.month
sales['quarter'] = sales['orderdate'].dt.quarter
sales['day_name'] = sales['orderdate'].dt.day_name()
```

CUSTOMER FEATURE ENGINEERING

Customers Features

Create customer-level analytical features.

```python
customer.info()
```

```python
# Customers Full Name

customer['full_name'] = (customer['givenname'] + " " + customer['surname'])
customer['full_name'].head(10)
```

```python
# Extracting Customers Age

customer['birthday'] = pd.to_datetime(customer['birthday'])
```

```python
# Calculating Customers Age

customer['age'] = 2025 - customer['birthday'].dt.year
customer['age'].head(10)
```

```python
# Segmenting customers on basis of their Age.

customer['age_group'] = pd.cut(customer['age'], bins = [0, 25, 40, 60, 100],
    labels = ['young', 'adult', 'middleage', 'senior']
)

```

PRODUCT FEATURE ENGINEERING

Product Features
Create pricing and category features.

```python
product.info()
```

```python
# Binning an product ont the price basis.

product['price_category'] = pd.cut(
    product['price'],
    bins=[0, 50, 200, 500, 100000],
    labels=['Low', 'Medium', 'High', 'Premium']
)
```

```python
product['category'] = (
    product['categoryname']
    .str.strip()
    .str.title()
)
```

STORE FEATURE ENGINEERING

    Store Features
    Create operational and size-related features.

```python
# Validating Stores 

store['opendate'] = pd.to_datetime(store['opendate'])
store['opendate'].head(10)
```

```python
# Calculating store age

store['store_age'] = 2025 - store['opendate'].dt.year
store['store_age'].head(10)
```

```python
# Creating bins on the basis of store size 

store['store_size'] = pd.cut(
    store['squaremeters'],
    bins=[0, 500, 1500, 5000, 20000],
    labels=['Small', 'Medium', 'Large', 'Enterprise']
)
store['store_size'].head(10)
```

DATE FEATURE ENGINEERING 

Date Features 
Enhance calander analysis capablities.

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date['year'] = date['date'].dt.year
date['month'] = date['date'].dt.month
date['quarter'] = date['date'].dt.quarter
date['day_name'] = date['date'].dt.day_name()
```

SAVE FEATURES TABLE

Saving all engineered tables to PostgresSQL

```python
sales.to_sql("feat_sales", engine, if_exists="replace", index=False)

customer.to_sql("feat_customer", engine, if_exists="replace", index=False)

product.to_sql("feat_product", engine, if_exists="replace", index=False)

store.to_sql("feat_store", engine, if_exists="replace", index=False)

date.to_sql("feat_date", engine, if_exists="replace", index=False)
```
